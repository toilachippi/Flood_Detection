```python
import os
import shutil
from sklearn.model_selection import train_test_split
from google.colab import drive

# Đường dẫn tới thư mục chứa ảnh gốc
source_folder = '/gdrive/MyDrive/Dpl/data_source/new_data'

# Đường dẫn tới các thư mục đích
train_path = os.path.join(source_folder, 'train')
valid_path = os.path.join(source_folder, 'valid')
test_path = os.path.join(source_folder, 'test')

# Tạo các thư mục đích nếu chưa tồn tại
os.makedirs(train_path, exist_ok=True)
os.makedirs(valid_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Danh sách tất cả các ảnh trong thư mục nguồn
all_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f)) and f.endswith(('.jpg', '.png', '.jpeg','gif'))]

# Chia dữ liệu: 70% train, 20% validation, 10% test
train_files, temp_files = train_test_split(all_files, test_size=0.3, random_state=42)
valid_files, test_files = train_test_split(temp_files, test_size=(1/3), random_state=42)

# Hàm để di chuyển các tập tin vào các thư mục đích
def move_files(file_list, source, destination):
    for file_name in file_list:
        shutil.move(os.path.join(source, file_name), destination)

# Di chuyển các tập tin vào các thư mục đích
move_files(train_files, source_folder, train_path)
move_files(valid_files, source_folder, valid_path)
move_files(test_files, source_folder, test_path)

print("Ảnh đã được chia vào các thư mục train, valid, và test.")
print(f"Tổng số ảnh: {len(all_files)}")
print(f"Số ảnh trong train: {len(train_files)}")
print(f"Số ảnh trong valid: {len(valid_files)}")
print(f"Số ảnh trong test: {len(test_files)}")
```

```python
import pandas as pd
import shutil
import os

# Đọc file CSV
csv_file = '/gdrive/MyDrive/Dpl/devset_images_gt.csv'  # Đường dẫn tới file CSV của bạn
df = pd.read_csv(csv_file)

# Tạo thư mục nếu chưa tồn tại
label_0_dir = '/gdrive/MyDrive/Dpl/data_source/new_data/test/normal'
label_1_dir = '/gdrive/MyDrive/Dpl/data_source/new_data/test/flooding'

os.makedirs(label_0_dir, exist_ok=True)
os.makedirs(label_1_dir, exist_ok=True)

# Đường dẫn tới thư mục chứa các file ảnh gốc
source_dir = '/gdrive/MyDrive/Dpl/data_source/new_data/test'

# Duyệt qua các hàng trong dataframe và di chuyển các file ảnh
for index, row in df.iterrows():
    file_id = str(row['id'])
    label = row['label']

    src_path = os.path.join(source_dir, f"{file_id}.jpg")  # Giả sử file ảnh có đuôi .jpg
    if label == 0:
        dst_path = os.path.join(label_0_dir, f"{file_id}.jpg")
    else:
        dst_path = os.path.join(label_1_dir, f"{file_id}.jpg")

    # Kiểm tra file ảnh có tồn tại không rồi mới di chuyển
    if os.path.exists(src_path):
        shutil.move(src_path, dst_path)
    else:
        print(f"File {src_path} không tồn tại.")

print("Hoàn thành việc tách file ảnh.")
```
